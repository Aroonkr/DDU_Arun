```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import klib as k
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("advertising.csv")
```

```python
df
```

```python
df.head()
```

```python
df.isnull().sum()*100
```

```python
df.isnull().mean()*100
```

```python
df.drop(columns=['Newspaper'],inplace=True)
```

```python
df
```

```python
for i in df.columns:
    if df[i].dtype !="object":
        sns.boxplot(y=df[i],color="red")
        plt.show();
```

 ## Selecting The Tuture Here Only One Column

```python
x=df[['TV','Radio']].values
x
```

## Selecting The Target Column

```python
y=df[['Sales']].values
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
y
```

## Splitting Into Training And Testing Data

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Model Building

```python
model=LinearRegression()
```

## Fitting The Model

```python
model.fit(x_train,y_train)
```

## Coefficient

```python
model.coef_
```

## Intercept

```python
model.intercept_
```

## Training Accuracy 

```python
testing_score=model.score(x_train,y_train)*100
testing_score
```

## Testing Accuracy

```python
testing_score=model.score(x_test,y_test)*100
testing_score
```

## R2 score

```python
R2_score=r2_score(y_test,model.predict(x_test))*100
R2_score
```

## Mean Absolute Error

```python
mae=mean_absolute_error(y_test,model.predict(x_test))
mae
```

## Mean Squared Error

```python
mse=mean_squared_error(y_test,model.predict(x_test))
mse
```

## Predicting The Values

```python
y_pred = model.predict(x)
```

```python
y_pred
```

## Creating A DataFrame With Actual Value In Order To Compare Them

```python
OUTPUT =pd.DataFrame(zip(y,y_pred),columns=("ACTUAL","PREDICTED"),dtype=float)
OUTPUT.head()
```

```python
plt.figure(figsize=(12,6))
plt.title("PREDICTED VS ACTUAL DATA")
plt.xticks()
sns.scatterplot(data=OUTPUT,x="ACTUAL",y="PREDICTED");
```

```python
plt.figure(figsize=(12,6))
plt.title("PREDICTED VS ACTUAL DATA")
sns.lineplot(data=OUTPUT,x="ACTUAL",y="PREDICTED");
```
